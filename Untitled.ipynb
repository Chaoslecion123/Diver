## Load data for world module

```python
from saleor.world.models import CountryArea, City, CityArea

infile = open('./geodir-ubigeo-inei.csv')
content = infile.readlines()
infile.close()
content = [line.strip().strip(',').split(',') for line in content]

# Getting departaments
departaments = sorted(set(list(zip(*content[1:]))[2]))

for departament_name in departaments:
    departament = CountryArea(country='PE', name=departament_name)
    departament.save()

# Getting provinces
departament = None
provinces = set(zip(*list(zip(*content[1:]))[1:]))
provinces = sorted(list(provinces), key=lambda x: (x[1], x[0]))

for province_name, departament_name in provinces:
    if departament is None or departament.name != departament_name:
        departament = CountryArea.objects.get(name=departament_name)  
    province = City(country_area=departament, name=province_name)
    province.save()

# Getting districts
city = None
districts = set(zip(*list(zip(*content[1:]))[:2]))
districts = list(districts)
districts = sorted(districts, key=lambda x: (x[1], x[0]))

for district_name, city_name in districts:
    if city is None or city.name != city_name:
        city = City.objects.get(name=city_name)  
    district = CityArea(city=city, name=district_name)
    district.save()
```

```python
from math import sqrt
from matplotlib import pyplot
from shapely.geometry import Polygon
from descartes.patch import PolygonPatch

# from figures import BLUE, SIZE, set_limits, plot_coords, color_isvalid
BLUE = '#6699cc'
GRAY = '#999999'
GM = (sqrt(5)-1.0)/2.0
W = 16.0
H = W*GM
SIZE = (W, H)

def plot_coords(ax, ob, color=GRAY, zorder=1, alpha=1):
    x, y = ob.xy
    ax.plot(x, y, 'o', color=color, zorder=zorder, alpha=alpha)

fig = pyplot.figure(1, figsize=SIZE, dpi=90)

# 1: valid polygon
ax = fig.add_subplot(121)

lim_polygons = []
great_polygon = None

for i, polygon in enumerate(woking_area_lim['polygons']):
    if len(polygon) > 2:
        polygon = Polygon(polygon)
        lim_polygons.append(Polygon(polygon))

        plot_coords(ax, polygon.exterior)
        patch = PolygonPatch(polygon, facecolor=BLUE, edgecolor=BLUE, alpha=0.5, zorder=2)
        ax.add_patch(patch)

pyplot.show()
```
